# Problem 1

We have that the joint distribution of two random variables $X$ and $Y$ is given by $f(x, y) = c\cdot x\exp(-y), 0 < x < y< \infty$ for some constant $c > 0$.

### Part 1

We need to find constant $c$ such that $f$ is a proper distribution function - We require that for $f$ to be a proper distribution function, it must integrate to 1 when integrated over its domain of definition

\begin{align}
    \int_0^{\infty} \int_0^y f(x, y) dx dy &= 1 \\
    \int_0^{\infty} \int_0^y c\cdot x\exp(-y) dx dy &= 1 \\
    c \int_0^{\infty} \exp(-y) (\int_0^y  x dx) dy &= 1 \\
    \frac{c}{2} \int_0^{\infty} \exp(-y) y^2 dy &= 1 \\
    \frac{c}{2} \cdot 2 = 1 \implies c &= 1
\end{align}

### Part 2

Marginal density $f_X(x)$ and $f_Y(y)$ is given by

\begin{align}
    f_X(x) &= \int_x^{\infty} f(x, y) dy = x \int_x^{\infty} e^{-y} dy = x e^{-x} \\
    f_Y(y) &= \int_0^y f(x, y) dx = e^{-y} \int_0^y x dx = \frac{y^2}{2} e^{-y}
\end{align}

### Part 3

We have that the conditional density of X given Y is

\begin{align}
    f_{X|Y}(x|y) = \frac{f(x, y)}{f_Y(y)} = \frac{xe^{-y}}{\frac{y^2}{2} e^{-y}} = \frac{2x}{y^2}
\end{align}

We have that the above won't be valid when $y = 0$ or when $y \not{\to} \infty$. Therefore, the domain is given by $0 < x < y < \infty$

### Part 4

\begin{align}
    \mathbb{E}[X] &= \int_0^{\infty} x f_X (x) dx = \int_0^{\infty} x^2 e^{-x} dx  = 2\\
    \mathbb{E}[Y] &= \int_0^{\infty} y f_Y (y) dy = \frac{1}{2} \int_0^{\infty} y^3 e^{-y} dy = 3\\
    \mathbb{E}[XY] &= \int_0^{\infty} \int_0^y xy f(x, y) dx dy = \int_0^{\infty} \int_0^y yx^2 e^{-y} dx dy = 8 \\
    \mathrm{Cov}[XY] &= \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y] = 8 - 6 = 2
\end{align}

### Part 5

By looking at a plot of $x < y$ and $x < \sqrt{y}$, we have that the required integration is given by

\begin{align}
\mathbb{P}(0 < X < \sqrt{Y}) = \int_0^{1} \int_0^y x\exp(-y) + \int_1^{\infty} \int_1^{\sqrt{y}} x\exp(-y) = 1 - \frac{5}{2e} + \frac{1}{e} = 1 - \frac{3}{2e}
\end{align}

# Problem 2

We let $y$ be the number of heads in $n$ tosses of a coin, whose probability of heads is $\theta$. Thus, we have that $y \sim Bin(n, \theta)$

### Part 1

We have that the prior distribution for $\theta$ is uniform on the range $[0, 1]$. Let $p(\theta) = 1$ for $0 \leq \theta \leq 1$.

\begin{align}
    p(y = k) = \int_0^1 p(y = k | \theta) p(\theta) d \theta = \int_0^1 p(y = k | \theta) d \theta = \int_0^1 {n \choose k} \theta^k (1 - \theta)^{n - k} d \theta
\end{align}

This is the kernel for the beta distribution with $\alpha = k + 1, \beta = n - k + 1$.

\begin{align}
    \implies p(y = k) = {n \choose k} \frac{\Gamma(k + 1) \Gamma(n - k + 1)}{\Gamma (n + 2)} = \frac{1}{n + 1}
\end{align}

### Part 2

\begin{align}
\theta &\sim Beta(\alpha, \beta) \implies p(\theta) \propto \theta^{\alpha - 1} (1 - \theta)^{\beta - 1} \\
y | \theta &\sim Bin(n, \theta) \implies p(y|\theta) \propto \theta^y (1 - \theta)^{n - y}. \\
\implies p(\theta | y) &\propto \theta^{y + \alpha - 1} (1 - \theta)^{n - y + \beta - 1} \propto Beta(\theta|\alpha + y, \beta + n - y). \\
\implies \mathbb{E}[\theta | y] &= \frac{\alpha + y}{\alpha + \beta + n} = \frac{\alpha}{\alpha + \beta} \cdot \left(\frac{\alpha + \beta}{\alpha + \beta + n}\right) + \frac{y}{n} \cdot \left(\frac{n}{\alpha + \beta + n}\right)
\end{align}

### Part 3

\begin{align}
    p(\theta | y) \propto \theta^y ( 1 - \theta)^{n - y} \propto Beta(\theta|y + 1, n - y + 1). \\
    \implies Var(\theta | y) = \frac{(y + 1)(n - y + 1)}{(n + 2)^2(n + 3)} \leq \frac{(\frac{n}{2} + 1)(n - \frac{n}{2} + 1)}{(n + 2)^2 (n + 3)} \leq \frac{1}{16} < \frac{1}{12} = Var(\theta)
\end{align}

### Part 4

We have that

\begin{align}
    p(\theta) &\propto \theta^{\alpha - 1} (1 - \theta)^{\beta - 1} \quad \mathrm{and} \quad Var(\theta) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)} \\
    p(\theta | y) &\propto \theta^{y + \alpha - 1} (1 - \theta)^{n - y + \beta -1} \propto Beta(\theta|\alpha + y, \beta + n - y) \\
    Var(\theta | y) &= \frac{(\alpha + y)(\beta + n - y)}{(\alpha + \beta + n)^2(\alpha + \beta + n + 1)}.
\end{align}

Suppose we consider the parameters, $n = 3, y = 2, \alpha = 2, \beta = 11$, we have that $Var(\theta) = \frac{11}{1183} < \frac{3}{272} = Var(\theta | y)$

```python
# Problem 3:
import jax.numpy as np
from jax import random, jit, vmap, grad, jacfwd
from jax.scipy.stats import multivariate_normal
from jax.tree_util import tree_multimap
from jax.experimental import optimizers

import numpy as onp
from scipy.stats import gaussian_kde
from IPython.display import clear_output
import pylab as pl

# Not sure if the expectation was to write a function
# that samples from the Gaussian dist without the direct function:
# Written this as an alternative option:
def custom_multivariate_normal(rng_key, mu, Sigma, num_samples):
    # Adding small value for numerical stability
    eps   = 1e-6
    Sigma = Sigma + eps * np.eye(mu.size)
    L     = np.linalg.cholesky(Sigma)

    return (  mu.reshape(1, mu.size) 
            + (L @ random.normal(rng_key, (mu.size, num_samples))).T)

# Distribution of X:
mu    = np.array([3, -1])
Sigma = np.array([[1.2, 0.4],
                  [0.4, 1.0]])

# Setting values for the variables alpha and beta:
alpha = 1.5
beta  = 0.2


# Number of Samples Considered:
num_samples = 5000
# Generating realizations of X:
X = onp.random.multivariate_normal(mu, Sigma, (num_samples,))
# Applying the transformation to Y
y1 = alpha * X[:, 0]
y2 = X[:, 1] / alpha + beta * (X[:, 0]**2 + alpha**2)
Y = onp.vstack([y1, y2])

# Performing KDE:
density = gaussian_kde(Y)
# Calculating the mean:
ymean = np.mean(Y, 1)
print("The mean vector is given by", ymean)
```

```python
# Setting Grid For Plotting:
xmin = Y[0, :].min() - 0.1
xmax = Y[0, :].max() + 0.1
ymin = Y[1, :].min() - 0.1
ymax = Y[1, :].max() + 0.1

# Plotting:
xgrid        = np.linspace(xmin, xmax, 500)
ygrid        = np.linspace(ymin, ymax, 500)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z            = density.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
pl.figure(dpi = 150)
pl.imshow(Z.T.reshape(Xgrid.shape),
          origin='lower', aspect='auto',
          extent=[xmin, xmax, ymin, ymax],
          cmap='jet')

cb = pl.colorbar()
cb.set_label("Density")

# Plotting the points of the distribution with some transparency
pl.plot(Y[0, :], Y[1, :], 'ko', alpha = 0.5, markersize = 1)
```

# Problem 4

We have the univariate normal distributions, $p(x) = \mathcal{N}(\mu_1, \sigma_1^2)$ and $q(x) = \mathcal{N}(\mu_2, \sigma_2^2)$

### Part 1

We need to compute the KL divergence between the two distributions analytically. For distributions $P$ and $Q$, relative entropy is given by the integral

\begin{align}
    D_{\mathrm{KL}} (P || Q) &= \int_{-\infty}^{\infty} p(x) \log \left(\frac{p(x)}{q(x)}\right) dx \\
    \implies D_{\mathrm{KL}} (P || Q) &= \int_{-\infty}^{\infty} p(x) (\log (p(x)) - \log (q(x)))dx \\
    &= \int_{-\infty}^{\infty} \left[ -\frac{1}{2} \log(2\pi) - \log(\sigma_1) - \frac{1}{2} \left(\frac{x-\mu_1}{\sigma_1}\right)^2 + \frac{1}{2}\log(2\pi) + \log(\sigma_2) + \frac{1}{2} \left(\frac{x-\mu_2}{\sigma_2}\right)^2  \right] \times \frac{1}{\sqrt{2\pi}\sigma_1} \exp\left[-\frac{1}{2}\left(\frac{x-\mu_1}{\sigma_1}\right)^2\right] dx \\
    &= \int_{-\infty}^{\infty} \left\{\log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{1}{2} \left[ \left(\frac{x-\mu_2}{\sigma_2}\right)^2 - \left(\frac{x-\mu_1}{\sigma_1}\right)^2 \right] \right\} \times \frac{1}{\sqrt{2\pi}\sigma_1} \exp\left[-\frac{1}{2}\left(\frac{x-\mu_1}{\sigma_1}\right)^2\right] dx \\
    &= \mathbb{E}_{1} \left\{\log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{1}{2} \left[ \left(\frac{x-\mu_2}{\sigma_2}\right)^2 - \left(\frac{x-\mu_1}{\sigma_1}\right)^2 \right]\right\} \\
    &= \log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{1}{2\sigma_2^2} \mathbb{E}_1 \left\{(X-\mu_2)^2\right\} - \frac{1}{2\sigma_1^2} \mathbb{E}_1 \left\{(X-\mu_1)^2\right\} \\
    &= =\log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{1}{2\sigma_2^2} \mathbb{E}_1 \left\{(X-\mu_2)^2\right\} - \frac{1}{2}
\end{align}

Now we have that

\begin{align}
    (X - \mu_2)^2 &= (X-\mu_1+\mu_1-\mu_2)^2 = (X-\mu_1)^2 + 2(X-\mu_1)(\mu_1-\mu_2) + (\mu_1-\mu_2)^2 \\
   \implies D_{\mathrm{KL}} &= \log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{1}{2\sigma_2^2}
\left[\mathbb{E}_1\left\{(X-\mu_1)^2\right\} + 2(\mu_1-\mu_2)\mathbb{E}_1\left\{X-\mu_1\right\} + (\mu_1-\mu_2)^2\right] - \frac{1}{2} \\
&=\log\left(\frac{\sigma_2}{\sigma_1}\right) + \frac{\sigma_1^2 + (\mu_1-\mu_2)^2}{2\sigma_2^2} - \frac{1}{2}
\end{align}

### Part 2

Assuming we have data from $p(x)$, we need to prove that minimizing the KL divergence to identify $\mu_2, \sigma_2$ is equivalent to performing MLE. We have that

\begin{align}
    \theta_{\min \mathrm{KL}} &= \arg\min_\theta D_{KL} (P || Q) \\
    &= \arg \min_\theta \mathbb{E}_{\mathbf{x} \sim p(\mathbf{x})} [\log p(\mathbf{x}) - \log q(\mathbf{x})]
\end{align}

But we have data for $p(\mathbf{x})$. So the first term above doesn't depend on $\theta$. Additionally, if the datapoints $\mathbf{x}$ are i.i.d, by the Law of Large Numbers, we have that

\begin{align}
    \theta_{\min \mathrm{KL}} &= \arg \min_\theta -\mathbb{E}_{\mathbf{x} \sim p(\mathbf{x})} \log q(\mathbf{x} | \theta) \\
    &= \arg \max_{\theta} \lim_{N \to \infty} \frac{1}{N} \sum_{i = 1}^N \log(q(x_i | \theta)) = \theta_{\mathrm{MLE}} 
\end{align}

### Part 3

Refer to the code below. We observe that upon decreasing the number of samples, the accuracy of the numerically estimated KL divergence decreases with an increase in error. Additionally, we'll note that the fit of the distribution becomes poorer to the data when decreasing number of samples as seen below.

```python
# Problem 4 (3):
mu_1 = 0
mu_2 = 1

sigma_1 = 1
sigma_2 = 1

def return_X_dists_KL(num_samples):
    # Creating realizations for the distributions:
    rng_key     = random.PRNGKey(0)
    key1, key2  = random.split(rng_key)
    X1          = mu_1 + sigma_1 * random.normal(key1, (num_samples,))
    X2          = mu_2 + sigma_2 * random.normal(key2, (num_samples,))

    # Performing Kernel Density Estimation
    p = gaussian_kde(X1)
    q = gaussian_kde(X2)

    # Computing KL divergence:
    from scipy.integrate import quad
    # Considering tolerances for float32
    kl_div = quad(lambda x: np.where(p(x) > 2e-7 and q(x) > 2e-7 , 
                                     p(x) * np.log(p(x) / q(x)), 0), 
                  -7, 7)[0]

    return X1, X2, p, q, kl_div

X1, X2, p, q, kl_div = return_X_dists_KL(1000)
print('Numerically Computed KL Divergence(1000 Samples):', kl_div)

# Now reducing number of samples:
_, _, _, _, kl_div1 = return_X_dists_KL(500)
_, _, _, _, kl_div2 = return_X_dists_KL(250)
X1_prime, X2_prime, p_prime, q_prime, kl_div3 = return_X_dists_KL(125)

print('Numerically Computed KL Divergence(500 Samples):', kl_div1)
print('Numerically Computed KL Divergence(250 Samples):', kl_div2)
print('Numerically Computed KL Divergence(125 Samples):', kl_div3)
```

```python
# Creating realizations for Z:
num_samples = 2000
rng_key     = random.PRNGKey(0)
Z           = random.normal(rng_key, (num_samples, 2))

# Defining the functions g and g_inv:
def g_inv(z):
    z1 = z[0]
    z2 = z[1]
    x1 = np.sign(z1) * abs(z1)**(1 / 3) + 1
    x2 = z2 - 0.5    
    return np.array([x1, x2])

def g(x, coeff_x1_basis, coeff_x2_basis):
    x1 = x[0]
    x2 = x[1]
    z1 = np.sum(x1**(np.arange(0, coeff_x1_basis.size)) * coeff_x1_basis)
    z2 = np.sum(x2**(np.arange(0, coeff_x2_basis.size)) * coeff_x2_basis)
    return np.array([z1, z2])

mu     = np.zeros(2)
Sigma  = np.eye(2)
logp_z = multivariate_normal.logpdf(Z, mean = mu, cov = Sigma)
```

```python
# Getting vmapped versions for g and g_inv:
g_inv_vec = vmap(g_inv, in_axes = (0,))
g_vec     = vmap(g, in_axes = (0, None, None))

# Generating pairs of X:
X = g_inv_vec(Z)
# Checking that the given forward and inverse transforms are consistent:
coeff1_true = np.array([-1, 3, -3, 1])
coeff2_true = np.array([0.5, 1])
assert(np.mean(abs(g_vec(X, coeff1_true, coeff2_true) - Z)) < 5e-7)
```

```python
J       = jacfwd(g)
# Getting vmapped versions of J and det:
J_vec   = vmap(J, in_axes = (0, None, None))
det_vec = vmap(np.linalg.det)
# logp_x  = (  multivariate_normal.logpdf(g_vec(X), mean = mu, cov = Sigma)
#            + np.log(np.abs(det_vec(J_vec(X)))))
```

```python
# Framing the gradient ascent problem as a gradient descent:
@jit
def loss(params, batch):
    coeff1, coeff2 = params
    return -np.mean(multivariate_normal.logpdf(g_vec(batch, coeff1, coeff2), 
                                               mean = mu, cov = Sigma) + 
                    np.log(np.abs(det_vec(J_vec(batch, coeff1, coeff2)))))
    
# Computing Gradients:
grad_loss = jit(grad(loss))
```

```python
# Running this block will show a visualization of progress for p(x)
# We assume polynomial basis P_3 for z1:
coeff1_est = random.normal(rng_key, (4,))
# We assume polynomial basis P_1 for z1:
coeff2_est = random.normal(rng_key, (2,))

params = coeff1_est, coeff2_est
batch  = X

# Gradient descent
# Learning Rate
alpha = 0.005
n_its = 10000

# Variables for Plotting:
x    = np.linspace(-5.0, 5.0, 401)
y    = np.linspace(-5.0, 5.0, 401)
X, Y = np.meshgrid(x, y)
grid = np.hstack([X.reshape(-1, 1), Y.reshape(-1, 1)])

loss_history = [loss(params, batch)]
frames = []
for i in range(n_its):
    g = grad_loss(params, batch)
    params = tree_multimap(lambda x, g: x - alpha * g, params, g)

    # Track progress:
    loss_history = loss_history + [loss(params, batch)]

    # Print Progress:
    if i % 100 == 0:
        clear_output(wait=True)
        print('Iteration:', i)
        print('Loss:', loss(params, batch))

        coeff1_est, coeff2_est = params
        Z = np.exp(  multivariate_normal.logpdf(g_vec(grid, coeff1_est, coeff2_est), 
                                                mean = mu, cov = Sigma) 
                    + np.log(np.abs(det_vec(J_vec(grid, coeff1_est, coeff1_est)))))

        Z = Z.reshape(X.shape)

        pl.figure(dpi = 150)
        pl.title("Iteration = {}, Loss = {}".format(i, loss(params, batch)))
        
        pl.hist2d(batch[:, 0], batch[:, 1], bins=50, range=[(-5,5), (-5,5)], density=True)
        pl.xlim(-5, 5)
        pl.ylim(-5, 5)        
        pl.contour(X, Y, Z, levels=10, lw=2)
        pl.show()
```
